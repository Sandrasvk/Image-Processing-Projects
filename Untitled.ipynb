### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from sklearn.utils import shuffle
from skimage.transform import resize
```

### Read all the files in the 3 folder

```python
mango = glob.glob('IMAGES/mango/*.JPEG')
orange = glob.glob('IMAGES/orange/*.JPEG')
cherry = glob.glob('IMAGES/cherry/*.JPEG')
```

```python
print(len(mango))
print(len(orange))
print(len(cherry))
```

### Read all images as pixel data and store it in a list along with its labels

### Read image data from all files and store it in a variable, Store the label as well

```python
image = []
label = []

for i in mango:
    img = plt.imread(i)
    image.append(img)
    label.append(0)

for i in orange:
    img = plt.imread(i)
    image.append(img)
    label.append(1)

for i in cherry:
    img = plt.imread(i)
    image.append(img)
    label.append(2)
```

### Storelabel value-name in a dictionary

```python
label_names = {'mango' : 0, 'orange' : 1, 'cherry' : 2}
```

### plotting

```python
plt.figure(figsize = (10,8))
for i in range(12):
    plt.subplot(3,4,i+1)
    plt.imshow(image[i])
plt.tight_layout()
```

### Shuffle images and labels

```python
image,label = shuffle(image,label)
```

### Plot the images once again

```python
plt.figure(figsize = (15,12))
for i in range(15):
    plt.subplot(4,4,i+1)
    plt.imshow(image[i])
plt.tight_layout()
```

### Shape of some images

```python
for i in range(10):
    print(image[i].shape)
```

### Resize all the images to a same height and width

```python
X = []

for i in image:
    X.append(resize(i,(400,400,3)))
```

### Convert images and labels to numpy array and store them as X and y variables

```python
X = np.array(X)
```

```python
y = np.array(label)
```

### Reshape the images to 1 dimension

```python
d1,d2,d3,d4 = X.shape
```

```python
X = X.reshape((d1, d2*d3*d4))
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 100)
```

### Create any model  I have used here a Random Forest Model and train it

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
%time model.fit(X_train,y_train)
```

### Make predictions with X_test

```python
%time y_pred = model.predict(X_test)
```

### Evaluate the Model

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred), annot = True, fmt = 'd')
```

```python
print(metrics.classification_report(y_test,y_pred))
```
